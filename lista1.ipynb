## Zadanie 1 lista 1 (regresja liniowa)




```python
import numpy as np
import matplotlib.pyplot as plt

xs = np.arange(0, 10, 0.1).reshape(-1, 1)
a1 = 5
a0 = 1
interferences = np.random.normal(2, 2, len(xs))
cnt = 0
ys = []
for x in xs:
    ys.append(a1*x + a0 + interferences[cnt])
    cnt += 1

ys = np.array(ys).reshape(-1, 1)
plt.scatter(xs, ys, alpha=0.5)

xs_with_ones = np.column_stack([np.ones((xs.shape[0],1)), xs])
```

Do danych zaproponuję model liniowy $y = a_{1}x + a_0$.\
Algorytm uczenia: $a = Y^TX(X^TX)^{-1}$\
Algorytm predykcji: $\overline{Y} = Xa^T$


```python
a = ys.T @ xs_with_ones @ np.linalg.inv(xs_with_ones.T @ xs_with_ones)
print(f"a0 = {a[0][0]}")
print(f"a1 = {a[0][1]}")

ys_pred = xs_with_ones @ a.T
plt.scatter(xs, ys, alpha=0.5)
plt.plot(xs, ys_pred, color="magenta")
```

# Zadanie 2 lista1 (regresja wielomianowa)



```python
import numpy as np
import matplotlib.pyplot as plt

xs = np.arange(-3, 3, 0.1).reshape(-1, 1)
a2 = 5
a1 = 10
a0 = 3
interferences = np.random.normal(-6, 6, len(xs))
cnt = 0
ys = []
for x in xs:
    ys.append(a2*x**2 + a1*x + a0 + interferences[cnt])
    cnt += 1

ys = np.array(ys).reshape(-1, 1)
plt.scatter(xs, ys, alpha=0.5)

n = 20
rows = xs.shape[0]
cols = n

phi_x = np.zeros((rows, cols))
phi_x = np.column_stack([phi_x, np.ones((phi_x.shape[0], 1))])

for c in range(cols):
    for r in range(rows):
        phi_x[r][c] = xs[r]** n
    n -= 1

a = ys.T @ phi_x @ np.linalg.inv(phi_x.T @ phi_x)

for coef in range(len(a[0])):
    print(f"a{coef} = {a[0][coef]}")

ys_pred = phi_x @ a.T
plt.plot(xs, ys_pred, color="magenta")
```

Wybrany model: wielomian stopnia n <br>
$\sum_{i=1}^{n} a_i x^i$ <br> 
Do uczenia należy wykorzystać tzw. uogólniony model liniowy \
$a = Y^T \Phi{(X)}(\Phi{(X^T)}\Phi{(X)})^{-1}$

# Zadanie 3 lista 1 (regresja liniowa z regularyzacją)
