# Lista 2 zadanie 1 (interpolacja liniowa)



```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

xs = np.linspace(0.0, 4.0, 12)
ys = [-0.7, -0.6, -0.5, -0.3, -0.1, 0.4, 0.75, 0.9, 0.8, 0.4, -0.6, -1]

f = interpolate.interp1d(xs, ys, kind="linear")
new_xs = np.linspace(0, 4, 68)
new_ys = f(new_xs)

plt.scatter(xs, ys)
plt.plot(xs, ys, color="orange")
plt.plot(new_xs, new_ys, "--", color="green")
```

# Zadanie 2 (interpolacja wielomianowa)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

# xs = np.linspace(0.0, 4.0, 4)
# print(xs)
# ys = [-0.7, 0.4, 0.75, -1]
xs = np.linspace(0.0, 4.0, 4)
ys = [-0.7, 0.4, 0.9, -1]

lagrange_interpolation = interpolate.lagrange(xs, ys)
# print(polynomia_interpolation)
new_xs = np.linspace(0, 4, 50)
new_ys = lagrange_interpolation(new_xs)

plt.scatter(xs, ys)
plt.plot(xs, ys, color="orange")
plt.plot(new_xs, new_ys, "--", color="green")
```

# zadanie 3 (Interpolacja funkcjami sklejanymi 3. stopnia)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

xs = np.linspace(0.0, 4.0, 12)
ys = [-0.7, -0.6, -0.5, -0.3, -0.1, 0.4, 0.75, 0.9, 0.8, 0.4, -0.6, -1]

cubic_spline_interpolation = interpolate.CubicSpline(xs, ys)
new_xs = np.linspace(0, 4, 50)
new_ys = cubic_spline_interpolation(new_xs)

plt.scatter(xs, ys)
plt.plot(xs, ys, color="orange")
plt.plot(new_xs, new_ys, "--", color="green")
```
