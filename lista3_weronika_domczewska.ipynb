# zadanie 1
 \
Normalizacja szeregów czasowych \
y = $\frac{x - min}{max - min}$

Po co i kiedy stosuje się normalizację szeregów czasowych?:
- gdy mamy do czynienia z danymi o różnej skali i chcemy je ze sobą porównać
- gdy chcemy stosować klasyfikację, np. algorytmem k najbliższych sąsiadów - wtedy odległość musi być wyrażona w jednej skali
- kiedy rozkład danych nie ma znaczenia


```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt

series = np.array(pd.date_range(start="2023-03-01", end="2023-03-07")).reshape(-1, 1)
temps = np.array([-1, -2.5, 0, 2.1, 3.9, 5, 6]).reshape(-1,1)

fig, ax = plt.subplots(2, 1, sharex=True)
plt.figure(figsize=(8, 6))
ax[0].plot(series, temps)
ax[0].set_title("Data before normalization")
normalizer = preprocessing.Normalizer(norm='l2')
normalized_temps = normalizer.transform(temps)
ax[1].plot(series, normalized_temps)
ax[1].set_title("Data after normalization")
plt.show()


```

# zadanie 2



Standaryzacja szeregów czasowych 

$y = \frac{x - \mu}{\sigma}$

- zmiana wartości średniej na 0 i odchylenia standardowego na 1
- zakłada, że standaryzowane dane są w przybliżeniu z rozkładu normalnego
- używane przy SVM (Support Vector Machines), regresji liniowej, regresji logistycznej
- potrzebna jest informacja o średniej i odchyleniu standardowym standaryzowanych danych


```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt

series = np.array(pd.date_range(start="2023-03-01", end="2023-03-07")).reshape(-1, 1)
temps = np.array([-1, -2.5, 0, 2.1, 3.9, 5, 6]).reshape(-1,1)

standarizer = preprocessing.StandardScaler()
standarized_temps = standarizer.fit_transform(temps)


fig, ax = plt.subplots(2, 1, sharex=True)
plt.figure(figsize=(8, 6))
ax[0].plot(series, temps)
ax[0].set_title("Data before standarization")

ax[1].plot(series, standarized_temps)
ax[1].set_title("Data after standarization")
plt.show()


```

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt

series = np.array(pd.date_range(start="2023-03-01", end="2023-03-07")).reshape(-1, 1)
temps = np.array([-1, -2.5, 0, 2.1, 3.9, 5, 6]).reshape(-1,1)

scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
scaled_temps = scaler.fit_transform(temps)

fig, ax = plt.subplots(2, 1, sharex=True)
plt.figure(figsize=(8, 6))
ax[0].plot(series, temps)
ax[0].set_title("Data before scaling")

ax[1].plot(series, scaled_temps)
ax[1].set_title("Data after scaling")
plt.show()
```
