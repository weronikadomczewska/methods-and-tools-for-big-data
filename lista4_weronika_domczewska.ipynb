Kurtoza jest miarą występowania danych odstających od wartości oczekiwanej, np. ile danych będzie na "ogonach" rozkładu normalnego. Może oznaczać tendencję danych do dużych zmian i odstępstw od normy. Używa się tej miary na przykład do oceny zmian cen produktów. 

Ujemna kurtoza wskazuje, na stosunkowo płaski rozkład - mało zróżnicowane względem średniej dane. 

```python
import numpy as np
import pandas as pd
import random
from scipy.stats import kurtosis

series = np.array(pd.date_range(start="2023-03-01", end="2023-03-31")).reshape(-1, 1)
n = 31
temps = np.zeros(n)
for i in range(n):
    temps[i] = random.randint(-5, 20)

print(f"Mean: {np.mean(temps)}")
print(f"Standard deviation: {np.std(temps)}")
print(f"Max value: {np.max(temps)}")
print(f"Min value: {np.min(temps)}")
print(f"Median: {np.median(temps)}")
print(f"Kurtosis: {kurtosis(temps)}")


```

wykładnik Hursta (Hurst exponent) - miara związana z "pamięcią" szeregu czasowego, czyli zależnościami statystycznymi pomiędzy danymi.
