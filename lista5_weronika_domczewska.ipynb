# Zadanie 1


PCA polega na wyszukaniu kierunkw, w ktrych wariancja zmiennych jest maksymalna oraz zrzutowaniu danych liniowo na now podprzestrze. Zadaniem jest zmniejszenie iloci wymiarw w zbiorze danych
- uywane w uczeniu nienadzorowanym (dla danych bez wczeniej nadanych etykiet)
- moe by uywane jako narzdzie wizualizacji zbiorw danych, filtr szumw, narzdzie do wydzielania cech


```python
import matplotlib.pyplot as plt
from sklearn import datasets, preprocessing
from sklearn.decomposition import PCA


def apply_pca_to_dataset(dataset):
    x = dataset.data
    y = dataset.target
    y = y.astype("int")

    x = preprocessing.StandardScaler().fit_transform(x)
    X_reduced = PCA(n_components=2).fit_transform(x)

    plt.figure(figsize=(8, 6))
    plt.scatter(
        X_reduced[:, 0], 
        X_reduced[:, 1],  
        c=y
    )
    plt.show()


iris = datasets.load_iris()
wine = datasets.load_wine()
diabetes = datasets.load_diabetes()

apply_pca_to_dataset(iris)
apply_pca_to_dataset(wine)
apply_pca_to_dataset(diabetes)
```
